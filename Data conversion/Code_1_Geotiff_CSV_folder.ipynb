```python
import os
from osgeo import gdal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to convert GeoTIFF to CSV
def geotiff_to_csv(input_geotiff, output_csv):
    # Open GeoTIFF file
    ds = gdal.Open(input_geotiff)
    if ds is None:
        print(f"Could not open the GeoTIFF file: {input_geotiff}")
        return

    # Translate GeoTIFF to XYZ
    xyz = gdal.Translate(output_csv.replace('.csv', '.xyz'), ds)

    # Close the GeoTIFF file
    ds = None

    # Read XYZ file into DataFrame
    df = pd.read_csv(output_csv.replace('.csv', '.xyz'), delim_whitespace=True, header=None, names=["x", "y", "snow cover"])
    #df = pd.read_csv(output_csv.replace('.csv', '.xyz'), delim_whitespace=True, header=None, names=["x", "y", "precipitation"])
    #df = pd.read_csv(output_csv.replace('.csv', '.xyz'), delim_whitespace=True, header=None, names=["x", "y", "temperature"])
    #df = pd.read_csv(output_csv.replace('.csv', '.xyz'), delim_whitespace=True, header=None, names=["x", "y", "windspeed"])
    #df = pd.read_csv(output_csv.replace('.csv', '.xyz'), delim_whitespace=True, header=None, names=["x", "y", "elevation"])

    # Write DataFrame to CSV
    df.to_csv(output_csv, index=False)

# Directory containing GeoTIFF files
input_directory = "Fieschertal\\Snow_Cover_Fieschertal\\"

# Output directory for CSV files
output_directory = "Fieschertal\\Snow_Cover_Fieschertal_csv\\"

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Process each GeoTIFF file
for filename in os.listdir(input_directory):
    if filename.endswith(".tif"):
        input_tiff_path = os.path.join(input_directory, filename)
        output_csv_path = os.path.join(output_directory, filename.replace('.tif', '.csv'))
        
        print(f"Converting {input_tiff_path} to CSV: {output_csv_path}")
        
        # Convert GeoTIFF to CSV
        geotiff_to_csv(input_tiff_path, output_csv_path)
```
