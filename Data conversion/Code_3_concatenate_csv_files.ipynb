```python
import os
from osgeo import gdal
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def concatenate_csv_files(input_directory, sorted_files):
    #ds = pd.read_csv(input_directory + os.listdir(input_directory)[0])
    ds = pd.read_csv(input_directory + sorted_files[0])
    for i in range(1, len(os.listdir(input_directory))):
        #df = pd.read_csv(input_directory + os.listdir(input_directory)[i])
        df = pd.read_csv(input_directory + sorted_files[i])
        ds = pd.concat([ds, df], axis=0)
    
    return ds

# Directory containing GeoTIFF files
input_directory = "Fieschertal\\WindSpeed_Fieschertal_csv1\\"
dp = os.listdir(input_directory)
sorted_files = sorted(dp, key=lambda x: (int(x.split('_')[0]), int(x.split('_')[1].split('.')[0])))

# Output directory for CSV files
output_directory = "Fieschertal\\combined\\"

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

result = concatenate_csv_files(input_directory, sorted_files)
#result.to_csv(output_directory + "1snow_cover_combined.csv", index = False)
#result.to_csv(output_directory + "2rainfall_combined.csv", index = False)
#result.to_csv(output_directory + "3temp_combined.csv", index = False)
result.to_csv(output_directory + "4windspeed_combined.csv", index = False)
```
