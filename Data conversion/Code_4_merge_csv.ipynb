```python
import os
import pandas as pd
import numpy as np
```

```python
def merge_csv_files(input_directory):
    common_columns = ['x', 'y', 'year', 'month']
    ds = pd.read_csv(input_directory + os.listdir(input_directory)[0])
    for i in range(1, len(os.listdir(input_directory))):
        df = pd.read_csv(input_directory + os.listdir(input_directory)[i])
        ds = pd.merge(ds, df, on = common_columns, how = 'inner')
    
    return ds

input_directory = "Fieschertal\\combined\\"
result = merge_csv_files(input_directory)
df = result[['x', 'y', 'year', 'month', 'snow cover', 'precipitation', 'temperature',
       'windspeed']]
df.to_csv("Fieschertal\\combined\\merged.csv", index = False)
```
