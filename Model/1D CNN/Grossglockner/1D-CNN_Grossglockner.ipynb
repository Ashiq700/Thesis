```python
# Import libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

import lime
import lime.lime_tabular
from lime.lime_tabular import LimeTabularExplainer

# Load Data
data = pd.read_csv("D:\\GEE\\Grossglockner\\combined\\allmerged.csv")

# Rename columns x and y to longi and lati
data = data.rename(columns={'x': 'longi', 'y': 'lati'})

print("Datasets:\n\n",data.head(10))
print(f'\n\nSize of the dataset: {data.shape}')

# Create new column 'new_label' where snow cover value <= 20 will be considered as 0 and others 1
data['new_label'] = data['snow cover'].apply(lambda x: 0 if x <= 20 else 1 )

# Input features
X = data.drop(columns=['snow cover','new_label'])

# level features
y = data['new_label'] 

# Perform Min-Max scaling on X
scaler = MinMaxScaler(feature_range=(0,1))
X_sc = scaler.fit_transform(X)

# Reshape data to fit 1D CNN (samples, time steps, features)
Xr = X_sc.reshape((X_sc.shape[0], X_sc.shape[1], 1))

#categorical classification
y_c = to_categorical(y, 2)

#split the dataset into train and test samples
X_train, X_test, y_train, y_test = train_test_split(Xr, y_c, test_size = 0.25, random_state = 42)

print(f"Shape of X_train: {X_train.shape}, Shape of X_test: {X_test.shape}, Shape of y_train: {y_train.shape}, Shape of y_test: {y_test.shape}")
print("\n\n")
# Define the model for categorical_crossentropy
model = Sequential([
    Input(shape = (X_train.shape[1], X_train.shape[2])),
    Conv1D(filters = 128, kernel_size = 2, activation = 'relu'),
    Conv1D(filters = 64, kernel_size = 2, activation = 'relu'),
    MaxPooling1D(pool_size = 2),
    Conv1D(filters = 32, kernel_size = 2, activation = 'relu'),
    MaxPooling1D(pool_size = 2),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(2, activation = 'softmax')  # For binary classification
])

model.summary()

# Compile the model
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs = 15, batch_size = 5600, validation_split = 0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'\n\nTest Accuracy: {accuracy*100:.2f}%')

# Training history
metrics = pd.DataFrame(history.history)

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.plot(metrics['loss'], label='Training Loss')
plt.plot(metrics['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Plot training & validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(metrics['accuracy'], label='Training accuracy')
plt.plot(metrics['val_accuracy'], label='Validation accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Make predictions
predictions = model.predict (X_test)

# Convert probabilities to binary output
binary_predictions = (predictions >= 0.5).astype(int)

# Covert categorical variables to one-hot vectors.
y_testNC = [ np.argmax(t) for t in y_test ]
y_predictNC = [ np.argmax(t) for t in binary_predictions ]

# Print classification report
cr = classification_report(y_testNC, y_predictNC, target_names=['no snowcover', 'snowcover'])
print("\n\nClassification Report:\n\n", cr)

#confusion matrix
cm = confusion_matrix(y_testNC, y_predictNC)
print("\n\nConfusion matrix:\n\n", cm)
print("\n")

# visualize confusion matrix
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ['{0:0.0f}'.format(value) for value in
                cm.flatten()]
group_percentages = ['{0:.2%}'.format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(cm.shape[0], cm.shape[1])

sns.heatmap(cm, annot=labels, fmt='', cmap="crest",
           xticklabels=['no snowcover', 'snowcover'],
           yticklabels =['no snowcover', 'snowcover'])

# display matrix
plt.ylabel('Actual',fontsize=12)
plt.xlabel('Prediction',fontsize=12)
plt.show()

print("\n\nLIME explanation")
# Flatten X_test for LIME
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Define LIME explainer
explainer = LimeTabularExplainer(
    X_test_flat, 
    feature_names = ['longi', 'lati', 'year', 'month', 'precipitation', 'temperature', 'windspeed', 'elevation'],
    class_names = ['No Snow', 'Snow'], 
    discretize_continuous = True
)

# Function to predict probabilities with the trained model
def predict_proba(x):
    x_reshaped = x.reshape(x.shape[0], X_train.shape[1], X_train.shape[2])
    return model.predict(x_reshaped)

# Choose an instance to explain
exp = explainer.explain_instance(X_test_flat[4], predict_proba, num_features=8)
exp.show_in_notebook(show_table=True)
```
