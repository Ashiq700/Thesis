### **Elevation in Grossglockner/Austria**

```python
#Import earth engine
import ee

#Trigger the authentication flow
ee.Authenticate()

#Initialize the library
ee.Initialize(project='modis-410817')
```

```python
# Import geemap
import geemap as emap
```

```python
# Define the region of interest (ROI)
roi = ee.Geometry.Rectangle([12.840, 46.983, 13.065, 47.082])

# Define the image collection
dataset = ee.Image('CGIAR/SRTM90_V4') \
    .select('elevation') \
    .clip(roi)
```

```python
# Visualization palette for temperature
vis = {
  'min': 100,
  'max': 5000,
  'palette': ['red', 'yellow', 'blue']
};
```

```python
#Create a center the map object
Map = emap.Map()
Map.center_object(roi, zoom=12)
# Add the layers to the map object.
Map.addLayer(dataset,vis,'Elevation')

# Draw the rectangle on the map
Map.addLayer(roi, {'color': 'red'}, 'roi')

#Add a layer control panel to the map
Map.addLayerControl()

#Display the map
Map
```

```python
# If the export has more than 1e8 pixels, set "maxPixels" higher.
task = ee.batch.Export.image.toDrive(
    image=dataset,
    description='Elevation_Grossglockner',
    folder='Elevation_Grossglockner',
    region=roi,
    scale=500,
    crs='EPSG:4326',
    maxPixels=1e13
)
task.start()
```
